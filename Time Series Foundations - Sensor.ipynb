## Import Packages and Data

```python
# Data from https://www.kaggle.com/datasets/nphantawee/pump-sensor-data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sensor_df = pd.read_csv("sensor.csv")
```

## Explore Data

```python
sensor_df.head()
```

```python
sensor_df.describe()
```

```python
sensor_df.isna().sum()
```

```python
sensor_df = sensor_df[['timestamp','sensor_03']]
```

```python
sensor_df.head()
```

## Create Timeseries Object

```python
sensor_df['timestamp'] = pd.to_datetime(sensor_df['timestamp'])
sensor_df.set_index('timestamp',inplace=True)
sensor_df.head()
```

```python
sensor_df = sensor_df.asfreq('h')
sensor_df.isna().sum()
```

(No missing values to impute)

## Plotting the Data

```python
sensor_df.plot(figsize=(20,5))
plt.title("Sensor 03 Trend",size = 20)
plt.show()
```

## Train-Test Splits

```python
size = int(len(sensor_df)*0.8)
sensor_train = sensor_df.iloc[:size]
sensor_test = sensor_df.iloc[size:]
```

```python
sensor_train.tail()
```

```python
sensor_test.head()
```

## Export Data

```python
sensor_train.to_csv("sensor - train.csv")
sensor_test.to_csv("sensor - test.csv")
sensor_df.to_csv("sensor - clean.csv")
```
